```python
import matplotlib.pyplot as plt

cKBS_parameters = ['10', '25', '50', '100']
cKBS_values = [2.20e-4, 2.29e-4, 3.16e-4, 3.92e-4]

FRN_parameters = ['10', '25', '50', '100']
FRN_values = [1.45e-3, 2.78e-3, 7.17e-3, 9.44e-2]

plt.figure(figsize=(6, 5), dpi= 300)
plt.plot(cKBS_parameters, cKBS_values, marker='s', label='LCDA')
plt.plot(FRN_parameters, FRN_values, marker='o', label='FRN')
plt.xlabel('Size of facts')
plt.ticklabel_format(axis='y', style='sci', scilimits=(2,4))
plt.grid(visible= True, linestyle = '--')
plt.ylabel('Execution Time')
#plt.title('Comparison of Methods')
plt.legend()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

# Set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# x and y data points
x = np.linspace(1, 100, 1000)
y = np.log(x)

plt.figure(figsize=(6, 5), dpi=300)
# logarithmic scale
plt.yscale('log')

# Plot the x and y data points
plt.plot(x, y, c="red", lw=3, linestyle="dashdot", label="y=log(x)")

# Place the legend
plt.legend()

# Display the plot
plt.show()
```
